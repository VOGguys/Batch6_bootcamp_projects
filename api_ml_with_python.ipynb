##**Basic ML-sklearn**

- build a ml model to predict the house sales in King County, Washington State, USA. The dataset consisted of historic data of houses sold between May 2014 to May 2015.

### Linear Regressions

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# read data
df = pd.read_csv("dataset/kc_house.csv")
```

```python
df
```

```python
df.info()
```

```python
# prepare data
X = df.drop(["id", "date","price"], axis=1)
y = df["price"]
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(
	X, y, test_size = 0.25, random_state = 42 #set.seed()
)
```

```python
# View shape
X_train.shape # ( row, column )
```

```python
# train model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# test model
p = model.predict(X_test)
print(p)
```

```python
#evaluated model 
model.score(X_test, y_test) # R2
```
